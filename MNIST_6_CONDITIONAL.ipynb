```python
import numpy as np
import matplotlib.pyplot as plt

import torch
from torch import optim
from torch.utils.data import DataLoader
import torch.nn as nn

from torchvision.transforms import transforms
import torchvision.datasets as datasets

from models import U_Net_big_cond
from RectifiedFlow import RectifiedFlow_Unet, train_cond

from util import load_MNIST, show_samples, show_trajectories, show_samples_array, show_sample_temperature, show_trajecrory_samples, show_sample_steps, show_sample_flows
```

```python
if torch.cuda.is_available():
    device = "cuda"
else:
    device = "cpu"
print(device)
```

```python
batch_size = 512
train_dataset, test_dataset, train_dataloader, test_dataloader, classes = load_MNIST(batch_size, classes=None)
input_shape = (batch_size, 3, 32, 32)

def get_samples(shape, device):
    return torch.randn(shape, device=device)
```

```python
v_model_1 = U_Net_big_cond(input_shape=input_shape, output_channels=1, device=device)
rectified_flow_1 = RectifiedFlow_Unet(v_model_1, device)
```

```python
optimizer_1 = torch.optim.Adam(rectified_flow_1.v_model.parameters(), lr=1e-3)
scheduler_1 = None # optim.lr_scheduler.ReduceLROnPlateau(optimizer_1, factor=0.8, patience=80, verbose=True)
```

```python
def default_network(out_features):
    hidden_num = 128
    network = nn.Sequential()

    network.append(nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1, padding=0))
    network.append(nn.BatchNorm2d(6))
    network.append(nn.ReLU())
    network.append(nn.MaxPool2d(kernel_size = 2, stride = 2))

    network.append(nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, stride=1, padding=0))
    network.append(nn.BatchNorm2d(16))
    network.append(nn.ReLU())
    network.append(nn.MaxPool2d(kernel_size = 2, stride = 2))

    network.append(nn.Flatten())

    network.append(nn.Linear(in_features=256, out_features=120))
    network.append(nn.ReLU())
    network.append(nn.Linear(in_features=120, out_features=84))
    network.append(nn.ReLU())
    network.append(nn.Linear(in_features=84, out_features=out_features))

    return network
```

```python
encoder = default_network(16).to(device)
encoder.load_state_dict(torch.load("condition_for_rf_mnist.pth"))
```

```python
rectified_flow_1 = train_cond(rectified_flow_1, optimizer_1, scheduler_1, train_dataloader, get_samples, encoder, device, epochs=600, noise_factor=0.01)
plt.plot(np.log(rectified_flow_1.loss_curve))
plt.show()
```

```python
def show_trajectories2(rectified_flow, rectified_flow_2, get_samples, num_samples, img_size, num_steps, x, y, device):
    rectified_flow.v_model.eval()
    plt.rcParams['figure.dpi'] = 120
    img_init = get_samples((num_samples, 1, img_size, img_size), device=device)
    img = rectified_flow.sample_ode(img_init, num_steps)
    s = torch.zeros(num_samples, len(img), 1, 32, 32)
    for i in range(len(img)):
        s[:, i] = img[i]
    for i in range(num_samples):
        plt.plot(s[i, :, 0, x, y].cpu().numpy())
    plt.ylim(-1.8, 1.8)
    plt.show()

    img = rectified_flow_2.sample_ode(img_init, num_steps)
    s = torch.zeros(num_samples, len(img), 1, 32, 32)
    for i in range(len(img)):
        s[:, i] = img[i]
    for i in range(num_samples):
        plt.plot(s[i, :, 0, x, y].cpu().numpy())
    plt.ylim(-1.8, 1.8)
    plt.show()
```

```python
show_trajectories2(rectified_flow_1, rectified_flow_reflow, get_samples, 40, 32, 100, 19, 14, device)
```
