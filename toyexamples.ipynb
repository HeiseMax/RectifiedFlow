```python
import torch

if torch.cuda.is_available():
    device = "cuda"
else:
    device = "cpu"
print(device)
```

```python
from toyproblems import Toy_problem1
toyexample = Toy_problem1(device)
toyexample.load()
if len(toyexample.flows) == 0:
    toyexample.train_flows(num_samples=5000, num_reflows=1, layers=2, batchsize=2048, iterations=10000)
    toyexample.save()
toyexample.show_flows(num_samples=5000, num_connections=100, num_steps=100)
```
